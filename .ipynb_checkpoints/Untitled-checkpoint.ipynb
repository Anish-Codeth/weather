```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("seattle-weather.csv")
df=df[df.columns[1:]]
df.head()
```

```python
df.shape
```

```python
df["weather"].unique()
```

```python
a=['low','high']
def handler(x):
    k=(x.max()-x.min()+1)/2
    mini=x.min()
#     print(a[int(((x-mini)/k).max())])
    return x.map(lambda y:a[int((y-mini)/k)])  

df_new=df[df.columns[:-1]].apply(handler,axis=0)
df_new['weather']=df['weather']
df_new
```

```python
indexes={}
indexes_array=[]
for x in df_new['weather'].unique():
    for y in df_new.columns:
        for z in a:
            indexes_array.append(x+"_"+y+'_'+z)
indexes_array
```

```python
drizzle=df_new[df_new['weather']=='drizzle'].shape[0]/df_new.shape[0]
print(drizzle)
```

```python
i=0
for y in df_new['weather'].unique():
    indexes[y]=df_new[df_new['weather']==y].shape[0]/df_new.shape[0]
    for z in df_new.columns:
        indexes[z+"_low"]=df_new[df_new[z]=='low'].shape[0]/df_new.shape[0]
        indexes[z+"_high"]=df_new[df_new[z]=='high'].shape[0]/df_new.shape[0]
        for x in a:
            df_drizzle=df_new[df_new['weather']==y]
            indexes[indexes_array[i]]=df_drizzle[df_drizzle['precipitation']==x].shape[0]/df_drizzle.shape[0]
            i=i+1
indexes
```

```python
which=['rain','drizzle_precipitation_low','drizzle_temp_max_low','drizzle_temp_min_low','drizzle_wind_low']
p=indexes[which[0]]
# for x in which:
#     y="_".join(which[1].split('_')[1:])
#     p=p*indexes[x]/indexes[y]
print(p*indexes[which[1]]*indexes[which[2]]/indexes['precipitation_low']/indexes['temp_max_low'])
```
